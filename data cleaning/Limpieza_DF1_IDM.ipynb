# **Primer dataset**

**La exploración y limpieza de este dataset nos ayudará a seleccionar y filtrar tipos y subtipos de delitos bajo el bien jurídico de la libertad personal en otros dataframes.**

# **1. Importación y exploración del dataset Incidencia Delictiva Municipal:**

```python
import numpy as np
import pandas as pd
import seaborn as sns


IDM=pd.read_csv("IDM_NM_abr25.csv", encoding='latin1')
display(IDM.head())
print("-----------------------------------")
print(IDM.shape)
print("-----------------------------------")
print(IDM.columns)
print("-----------------------------------")
print(pd.unique(IDM["Entidad"]))
```

### **1.1 Se renombran las columnas con caracteres especiales para que no haya problema con la manipulación de los datos:**

```python
IDM = IDM.rename(columns={
    'Bien jurídico afectado': 'Bien_jurídico_afectado',
    'Tipo de delito': 'Tipo_de_delito',
    'Subtipo de delito': 'Subtipo_de_delito'
})
display(IDM.head())
print(IDM.columns)
```

# **2. Filtrando el dataframe:**

### **2.1 Subsetting que abarca únicamente los datos para CDMX:**

```python
CDMX=IDM.query(' Entidad == "Ciudad de México" ')
display(CDMX.head())
print("-----------------------------------")
print(pd.unique(CDMX["Bien_jurídico_afectado"]))
print("-----------------------------------")
print(pd.unique(CDMX["Tipo_de_delito"]))
print("-----------------------------------")
print(pd.unique(CDMX["Subtipo_de_delito"]))
print("-----------------------------------")
print(pd.unique(CDMX["Modalidad"]))
```

### **2.2 Se filtra el dataframe para el bien jurídico afectado de la libertad personal, que es el tema central del proyecto:**

```python
PRUEBA_CDMX=IDM.query(' Entidad == "Ciudad de México" and Bien_jurídico_afectado == "Libertad personal" ')
display(PRUEBA_CDMX.head())
print("-----------------------------------")
print(PRUEBA_CDMX.shape)
print("-----------------------------------")
print(pd.unique(PRUEBA_CDMX["Tipo_de_delito"]))
print("-----------------------------------")
print(pd.unique(PRUEBA_CDMX["Subtipo_de_delito"]))
print("-----------------------------------")
print(pd.unique(PRUEBA_CDMX["Modalidad"]))
```

### **2.2 Más filtros para el dataframe:**

```python
#Se evalúa la posibilidad de eliminar algunas columnas y renglones que no aporten a las estadísticas de acuerdo a los objetivos del proyecto:

print((PRUEBA_CDMX["Tipo_de_delito"] == PRUEBA_CDMX["Subtipo_de_delito"]).value_counts())
print("-----------------------------------")
print(PRUEBA_CDMX['Clave_Ent'].dtype , PRUEBA_CDMX['Cve. Municipio'].dtype, PRUEBA_CDMX['Año'].dtype )
```

```python
#Se define el dataframe con el que se trabajará. La eliminación de la columna del año 2015 es para coincidir con el registro temporal de otros datasets:

IDM_CDMX_LibertadPersonal=IDM.query(' Entidad == "Ciudad de México" and (Bien_jurídico_afectado == "Libertad personal" and Año != 2015) ').drop([ \
    'Clave_Ent', 'Entidad', 'Tipo_de_delito', 'Subtipo_de_delito', 'Bien_jurídico_afectado'], axis=1)
    
display(IDM_CDMX_LibertadPersonal.head())
print("-----------------------------------")
print(pd.unique(IDM_CDMX_LibertadPersonal["Municipio"]))
print("-----------------------------------")
print(pd.unique(IDM_CDMX_LibertadPersonal["Año"]))
print("-----------------------------------")
print(IDM_CDMX_LibertadPersonal.columns)
print("-----------------------------------")
print(IDM_CDMX_LibertadPersonal.shape)
```

# **3. Dando formato al dataframe:**

Se transformará de formato ancho a formato largo, pues comunmente las bibliotecas de visualización de datos, como Seaborn, son más amigables con formatos largos.

```python
IDM_CDMX_LibertadPersonal=IDM_CDMX_LibertadPersonal.melt(id_vars=['Año', 'Municipio', 'Modalidad', 'Cve. Municipio']).rename(columns={ \
    'Municipio': 'Alcaldía', 'variable': 'Mes', 'value': 'Numero_de_registros'})
display(IDM_CDMX_LibertadPersonal.head())
print("-----------------------------------")
print(IDM_CDMX_LibertadPersonal.info())
print("-----------------------------------")
print(IDM_CDMX_LibertadPersonal.shape)
print("-----------------------------------")
print(IDM_CDMX_LibertadPersonal.isna().any())
print("-----------------------------------")
print(pd.unique(IDM_CDMX_LibertadPersonal["Año"]))
print("-----------------------------------")
print(pd.unique(IDM_CDMX_LibertadPersonal["Modalidad"]))
```

# **4. Imputación de valores NaN:**

Se imputarán los datos NaN interpretándolos como la ausencia de registros de secuestros, es decir, se van a rellenar con ceros.

De lo anterior, nos quedaremos con el dataframe en donde el número de registros son distintos de cero, pues con base a estos será el análisis.


```python
IDM_CDMX_LibertadPersonal=IDM_CDMX_LibertadPersonal.fillna(0).query(' Numero_de_registros != 0 ').sort_values(by=['Año','Alcaldía'], ascending=[True, True])
display(IDM_CDMX_LibertadPersonal.head())
print("-----------------------------------")
print(pd.unique(IDM_CDMX_LibertadPersonal["Alcaldía"]))
print("-----------------------------------")
print(IDM_CDMX_LibertadPersonal.info())
```

# **5. Cambiando el tipo de datos para el análisis:**

```python
#Cambiamos el formato float64 a int64 para la columna de número de registros:

IDM_CDMX_LibertadPersonal['Numero_de_registros'] = IDM_CDMX_LibertadPersonal['Numero_de_registros'].astype(int)
print(IDM_CDMX_LibertadPersonal.info())
print(IDM_CDMX_LibertadPersonal.shape)
display(IDM_CDMX_LibertadPersonal.head())
```
