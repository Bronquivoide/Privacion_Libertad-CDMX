# **Tercer dataset**

# **1. Importación y exploración inicial del dataset Cifras de Víctimas del Fuero Común:**

```python
import numpy as np
import pandas as pd

secuestros=pd.read_csv("IDVFC_NM_ene25.csv", encoding='latin1')
display(secuestros.head(2))

print(secuestros.columns)
print("-----------------------------------")
print(secuestros.shape)
print("-----------------------------------")
print(pd.unique(secuestros['Entidad']))
print("-----------------------------------")
print(pd.unique(secuestros['Bien jurídico afectado']))
```

### **1.1 Tratamiento de las columnas con nombres que contienen caracteres especiales:**

```python
secuestros=secuestros.rename(columns={
    'Bien jurídico afectado': 'Bien_juridico_afectado',
    'Tipo de delito': 'Tipo_de_delito',
    'Subtipo de delito': 'Subtipo_de_delito',
    'Rango de edad': 'Rango_de_edad'})

print(secuestros.columns)
```

# **2. Filtrando el dataframe para el bien jurídico afectado de la libertad personal en CDMX:**

Con base en la exploración de los primeros dos datasets, filtramos de la siguiente manera:

```python
secuestros=secuestros.query(' Entidad == "Ciudad de México" and Bien_juridico_afectado ==  "Libertad personal" ')
display(secuestros.head(2))

#Definiendo el dataframe:
print(pd.unique(secuestros['Tipo_de_delito']))
print("-----------------------------------")
print(pd.unique(secuestros['Subtipo_de_delito']))
print("-----------------------------------")
print(pd.unique(secuestros['Modalidad']))
```

```python
#Eliminando columnas innecesarias:
secuestros=secuestros.drop(['Clave_Ent', 'Entidad', 'Bien_juridico_afectado', 'Subtipo_de_delito'], axis=1)
display(secuestros.head(2))
print("-----------------------------------")
print(secuestros.columns)
```

# **3. Dando un formato largo al dataframe :**

```python
secuestros=secuestros.melt(id_vars=['Año', 'Tipo_de_delito', 'Modalidad', 'Sexo', 'Rango_de_edad'], var_name='Mes', value_name="Numero_registros")
display(secuestros.head(2))
print("-----------------------------------")
print(secuestros.columns)
```

# **4. Definiendo el dataframe :**

Nos quedaremos únicamente con las cuentas que importan para el análisis, que son los casos con números de registros distintos de cero. Para ello veremos primero los valores NaN.

### **4.1 Tratamiento de valores NaN:**

Estos se interpretan como una ausencia de registros, por lo que se asignarán a cero.

```python
print(secuestros.info())
print("-----------------------------------")
print(secuestros.isna().any())
print("-----------------------------------")
print(secuestros.isna().sum())
```

```python
secuestros['Numero_registros']=secuestros['Numero_registros'].fillna(0)
print(secuestros.isna().sum())
```

### **4.2 Filtrando para registros distintos de cero:**

```python
#Asignando un formato entero a los valores flotantes:
secuestros['Numero_registros']=secuestros['Numero_registros'].astype(int)

#Filtrando:
secuestros=secuestros.query(' Numero_registros != 0')

display(secuestros.head(2))
print("-----------------------------------")
print(secuestros.info())
```

# **5. Ordenando y filtrando por año y mes para coincidir con los otros datasets:**

Los años como datos de formato int se pueden ordenar fácilmente. Para los meses como strings hay que establecer un orden categórico.

```python
#Ordenando por año:
secuestros=secuestros.query('Año != 2015')

#Asignando el orden de los meses:
orden_meses=[ 'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
secuestros['Mes']=pd.Categorical(secuestros['Mes'], categories=orden_meses, ordered=True)

#Filtrando por año y por mes:
secuestros=secuestros.sort_values(by=['Año', 'Mes'], ascending=[True, True])
display(secuestros.head(10))
```

# **6. Dataframe final:**

```python
secuestros.to_csv('df3_VFC_Libertad.csv', index=False)
```
