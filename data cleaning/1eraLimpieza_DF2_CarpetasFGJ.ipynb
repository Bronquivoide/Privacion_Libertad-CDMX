```python
import numpy as np
import pandas as pd
import seaborn as sns


CARP=pd.read_csv("carpetasFGJ_acumulado_2025_01.csv", encoding='utf-8')
display(CARP.head())
print("-----------------------------------")
print(CARP.columns)
```

```python
df_10_17=pd.read_csv("carpetasFGJ_acumulado_2025_01.csv", encoding='latin1', usecols=[10, 17])
display(df_10_17.head())
print("-----------------------------------")
print(df_10_17.info())
```

```python
columnas_a_eliminar=CARP[['hora_inicio', 'anio_hecho', 'mes_hecho', 'competencia', 'colonia_hecho', 'alcaldia_catalogo', 'latitud',
       'longitud', 'municipio_hecho']]
```

```python
carpetasFGJ=CARP.drop(columns=columnas_a_eliminar)
#.drop(columns=cols_a_dropear)
display(carpetasFGJ.head())
print("-----------------------------------")
print(carpetasFGJ.columns)
print("-----------------------------------")
print(carpetasFGJ.shape)
print("-----------------------------------")
print(carpetasFGJ.info())
print("-----------------------------------")
print(carpetasFGJ.isna().any())
print("-----------------------------------")
print(pd.unique(carpetasFGJ['delito']))
```

```python
carpetasFGJ['delito_es_secuestro']=carpetasFGJ['delito'].str.contains('secuestro|fico|libertad|rapto|desapar', case=False)
display(carpetasFGJ.head())
print("-----------------------------------")
print(carpetasFGJ.columns)
print("-----------------------------------")
print(carpetasFGJ.info())
print("-----------------------------------")
print(pd.unique(carpetasFGJ['anio_inicio']))
```

```python
CARPETAS_SECUESTROS=carpetasFGJ.query(' delito_es_secuestro == True ')
display(CARPETAS_SECUESTROS.head())
print("-----------------------------------")
print(pd.unique(CARPETAS_SECUESTROS['delito']))
```

```python
secuestrosFGJ=CARPETAS_SECUESTROS.query(' delito != "TRAFICO DE INFLUENCIA" ').drop( 'delito_es_secuestro', axis=1)
display(secuestrosFGJ.head())
print("-----------------------------------")
print(secuestrosFGJ.columns)
print("-----------------------------------")
print(secuestrosFGJ.shape)
print("-----------------------------------")
print(secuestrosFGJ.info())
print("-----------------------------------")
print(secuestrosFGJ.isna().any())
print("-----------------------------------")
```
